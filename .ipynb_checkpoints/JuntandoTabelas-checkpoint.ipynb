## Juntando tabelas utilizando o pandas

##### Criei 2 tabelas de excel com os respectivos nomes loja 1 e 2.

##### Com 5 colunas e 10 valores ao total, sendo as colunas CODIGO_PRODUTO, DESCRICAO, QTDE_ITENS_VENDIDOS, TOTAL_VENDA e CODIGO_LOJA 

```python
import pandas as pd
```

```python
tabela1 = pd.read_excel("loja1.xlsx")
tabela2 = pd.read_excel("loja2.xlsx")
```

```python
# Mostrando apenas os 10 primeiros valores
tabela1.head(10)
```

```python
# Mostrando os 5 primeiros valores
tabela2.head(5)
```

## CONCAT

#### O metodo CONCAT serve para unir as 2 tabelas, ou seja, concatenar as tabelas de modo que as informacoes fique uma abaixo da outra


```python
tabelas = pd.concat([tabela1, tabela2])
tabelas.head(15)
```

### Usando o mesmo exemplo acima, vamos concatenas as 2 tabelas porem nesse caso vamos ignorar o indice de cada tabela e gerar uma nova sequencia de indices automaticamente

```python
tabelas = pd.concat([tabela1, tabela2], ignore_index = True)
tabelas.head(15)
```

## OUTRO EXEMPLO

```python
produtos = pd.read_excel("produtos.xlsx")
produtos.head(10)
```

# MERGE

### Uma outra forma de juntar tabelas e com o comando MERGE

### O comando MERGE e uma especie de PROCV do excel / Exatamente igual um JOIN do SQL

### por padrao o metodo utilizado o HOW = INNER

```python
# tabela_geral = pd.merge(tabelas, produtos, on = 'CODIGO_PRODUTO')
tabela_geral = pd.merge(tabelas, produtos, on = ['CODIGO_PRODUTO', 'DESCRICAO'])
tabela_geral.head(20)
```

### Podemos utilizar o parametro HOW = RIGHT para trazer todos os itens da tabela da direita.
### E os itens que nao forem encontrados na tabela da esquerda serao preenchidos como valor NULO

```python
tabela_geral = pd.merge(tabelas, produtos, on = 'CODIGO_PRODUTO', how = 'right')
tabela_geral.head(20)
```

### Podemos utilizar o parametro HOW = LEFT para trazer todos itens da tabela da esquerda e acrescentar os itens da tabela da direita

```python
tabela_geral = pd.merge(tabelas, produtos, on = 'CODIGO_PRODUTO', how = 'left')
tabela_geral.head(20)
```

### O OUTER ele vai pegar todos os valores das 2 tabelas e vai juntar em uma unica tabela.

### Os itens que nao houver correspondencia serao preenchidos com valor NULO

```python
tabela_geral = pd.merge(tabelas, produtos, on = 'CODIGO_PRODUTO', how = 'outer')
tabela_geral.head(20)
```
